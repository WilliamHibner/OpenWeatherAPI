```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import json
import time
from citipy import citipy

# Openweather API Key
from config import api_key
```

## Generate Cities List

```python
#step1 create random location from uniform numpy
# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)# Tested at 150 min queries to 500 
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)# change from 1500 to min queries to 500
```

```python
# step 2  loop over the random locations and extract city data from citipy 

cities=[]
for i in range(len(lats)):
    lat  = lats[i]
    lng  = lats[i]
    city        = citipy.nearest_city(lat, lng).city_name # cityPy
    countryCode = citipy.nearest_city(lat, lng).country_code # cityPy
    cities.append((city, countryCode)) # append to city list 
    
    
# output list of cities from random locations 
```

```python
#loop over city from above cell and request from weather location each city weather infom and save it in a list 

city_data = []
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + api_key 
counter=1
for city in cities:
    print(city)
    city_url = url + "&q=" + city[0] + ',' + city[1]
    #print(city_url)
    print(counter)
    print(f'City: {city[0]}')
    counter =  counter +1
    print('------')
    
    try:
        city_weather = requests.get(city_url).json() # extract data and save it as json 
        
        # Parse out the max temp, humidity, and cloudiness 
        # from the json extract the info 
        city_lat = city_weather["coord"]["lat"]
        city_lng = city_weather["coord"]["lon"]
        city_max_temp = city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]
        city_date = city_weather["dt"]
        
        # saving the extracted data in a dic 
        city_weather = {"Lat"       : city_lat,
                        "Lng"       : city_lng,
                        "Max Temp"  : city_max_temp,
                        "Humidity"  : city_humidity,
                        "Cloudiness": city_clouds,
                        "Wind Speed": city_wind,
                        "Date"      : city_date}

        # Append the City information into dictionary t0 city_data list
        
        city_data.append(city_weather)
        

    except:
        print("Skipping... City Not found...")
        pass
print('Done')
```

```python
# create a df from the city_data list 
df = pd.DataFrame(city_data)
```

```python
# Scatter plot for the selected coordinates  
plt.scatter(df.Lng, 
            df.Lat,
            edgecolor="red", linewidths=.15, marker="o", 
            alpha=0.5, label="Weather Cities")

# Incorporate the other graph properties
#plt.title("Selected LATS & LNGS")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.12),
          ncol=3, fancybox=True, shadow=True)
plt.grid(True)

# Show plot
plt.show()
```

```python

# Build scatter plot for latitude vs temperature
lats     =df.Lat
max_temps=df['Max Temp']
plt.scatter(lats, 
            max_temps,
            edgecolor="black", linewidths=1, marker="o", 
            alpha=0.38)

# Incorporate the other graph properties
plt.title("City Latitude vs. Max Temperature (%s)" % time.strftime("%x"))
plt.ylabel("Max Temperature (F)")
plt.xlabel("Latitude")
plt.grid(True)

# Save the figure
plt.savefig("Fig1.png")

# Show plot
plt.show()
```
